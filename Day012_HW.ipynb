```python
# ① 取出今天所有的發文
import requests
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text, 'html5lib')

new_list=[]
for d in soup.find(class_ = "part_list_2").find_all('h3'):
    print(d.find(class_='date').text, d.find(class_='tag').text, d.find_all('a')[0].text)
    new_list.append([d.find(class_='date').text, d.find(class_='tag').text, d.find_all('a')[0].text])
```

```python
# ② 如果想要依照類別分類，怎麼儲存會比較好？
import pandas as pd

columns = ['日期','類別','標題']

table = pd.DataFrame(new_list, columns=columns)
table
```

```python
# ③ 哪一個類別的文章最多

group = table.groupby('類別')
table2 = pd.DataFrame(group.size(), columns=['文章數量'])
table2.sort_values(by='文章數量', ascending=False)
```
