```python
# 1. Dcard 網址： https://www.dcard.tw/f

# Request 取回之後該怎麼取出資料，資料型態是什麼？
# 為什麼要使用 BeatifulSoup 處理？處理後的型態是什麼？
# 觀察一下知乎回來的資料好像有點怪怪的，該怎麼解決？
```

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.dcard.tw/f'
r= requests.get(url)
r.encoding = 'utf-8'
print("Request 取回之後該怎麼取出資料，資料型態是什麼？", type(r.text))

soup = BeautifulSoup(r.text,'html5lib')
print("為什麼要使用 BeatifulSoup 處理？處理後的型態是什麼？",type(soup))
```

```python
# 2. 知乎： https://www.zhihu.com/explore

# 觀察一下知乎回來的資料好像有點怪怪的，該怎麼解決？
```

```python
url = 'https://www.zhihu.com/explore'
r = requests.get(url)
r.encoding = 'utf-8'

print(r.text)

print("觀察一下知乎回來的資料好像有點怪怪的，該怎麼解決？")
headers = {'user-agent':'my-app/0.0.1'}
r = requests.get(url,headers=headers)
r.encoding = 'utf-8'
print(r.text)
```
